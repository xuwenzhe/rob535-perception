```python
import pandas as pd
import numpy as np
from ast import literal_eval

IMG_WIDTH = 1914
IMG_HEIGHT = 1052
```

```python
cocov1 = pd.read_csv('tf_features/cocov1-steps-201k-kaggle-test.tsv',sep='\t')

cocov1['cocov1-can1-label'] = cocov1['labels'].apply(lambda x : int(x[1]))
cocov1['cocov1-can1-score'] = cocov1['scores'].apply(lambda x : literal_eval(x)[0])
cocov1['cocov1-can1-ymin'] = cocov1['boxes'].apply(lambda x : IMG_HEIGHT*literal_eval(x)[0][0])
cocov1['cocov1-can1-xmin'] = cocov1['boxes'].apply(lambda x : IMG_WIDTH*literal_eval(x)[0][1])
cocov1['cocov1-can1-ymax'] = cocov1['boxes'].apply(lambda x : IMG_HEIGHT*literal_eval(x)[0][2])
cocov1['cocov1-can1-xmax'] = cocov1['boxes'].apply(lambda x : IMG_WIDTH*literal_eval(x)[0][3])
cocov1['cocov1-can1-box-width'] = cocov1['cocov1-can1-xmax'] - cocov1['cocov1-can1-xmin']
cocov1['cocov1-can1-box-height'] = cocov1['cocov1-can1-ymax'] - cocov1['cocov1-can1-ymin']
cocov1['cocov1-can1-box-ratio'] = cocov1['cocov1-can1-box-height']/cocov1['cocov1-can1-box-width']
cocov1['cocov1-can1-box-area'] = cocov1['cocov1-can1-box-height']*cocov1['cocov1-can1-box-width']


cocov1 = cocov1.drop(['labels','scores','boxes'],axis=1)
cocov1.head()
```

```python
kitti = pd.read_csv('tf_features/kitti-steps-660k-kaggle-test.tsv',sep='\t')

kitti['kitti-can1-label'] = kitti['labels'].apply(lambda x : int(x[1]))
kitti['kitti-can1-score'] = kitti['scores'].apply(lambda x : literal_eval(x)[0])
kitti['kitti-can1-ymin'] = kitti['boxes'].apply(lambda x : IMG_HEIGHT*literal_eval(x)[0][0])
kitti['kitti-can1-xmin'] = kitti['boxes'].apply(lambda x : IMG_WIDTH*literal_eval(x)[0][1])
kitti['kitti-can1-ymax'] = kitti['boxes'].apply(lambda x : IMG_HEIGHT*literal_eval(x)[0][2])
kitti['kitti-can1-xmax'] = kitti['boxes'].apply(lambda x : IMG_WIDTH*literal_eval(x)[0][3])

kitti['kitti-can1-box-width'] = kitti['kitti-can1-xmax'] - kitti['kitti-can1-xmin']
kitti['kitti-can1-box-height'] = kitti['kitti-can1-ymax'] - kitti['kitti-can1-ymin']
kitti['kitti-can1-box-ratio'] = kitti['kitti-can1-box-height']/kitti['kitti-can1-box-width']
kitti['kitti-can1-box-area'] = kitti['kitti-can1-box-height']*kitti['kitti-can1-box-width']

kitti = kitti.drop(['labels','scores','boxes'],axis=1)
kitti.head()
```

```python
incep = pd.read_csv('tf_features/incep-steps-200k-kaggle-test.tsv',sep='\t')

incep['incep-can1-label'] = incep['labels'].apply(lambda x : int(x[1]))
incep['incep-can1-score'] = incep['scores'].apply(lambda x : literal_eval(x)[0])
incep['incep-can1-ymin'] = incep['boxes'].apply(lambda x : IMG_HEIGHT*literal_eval(x)[0][0])
incep['incep-can1-xmin'] = incep['boxes'].apply(lambda x : IMG_WIDTH*literal_eval(x)[0][1])
incep['incep-can1-ymax'] = incep['boxes'].apply(lambda x : IMG_HEIGHT*literal_eval(x)[0][2])
incep['incep-can1-xmax'] = incep['boxes'].apply(lambda x : IMG_WIDTH*literal_eval(x)[0][3])

incep['incep-can1-box-width'] = incep['incep-can1-xmax'] - incep['incep-can1-xmin']
incep['incep-can1-box-height'] = incep['incep-can1-ymax'] - incep['incep-can1-ymin']
incep['incep-can1-box-ratio'] = incep['incep-can1-box-height']/incep['incep-can1-box-width']
incep['incep-can1-box-area'] = incep['incep-can1-box-height']*incep['incep-can1-box-width']

incep = incep.drop(['labels','scores','boxes'],axis=1)
incep.head()
```

```python
cocov1_kitti = cocov1.merge(kitti,how='left',on='guid/image',sort=True)
cocov1_kitti_incep = cocov1_kitti.merge(incep,how='left',on='guid/image',sort=True)
cocov1_kitti_incep.head()
```

```python
cocov1_kitti_incep.to_csv('lgb_test.tsv',sep='\t',index=False)
```
