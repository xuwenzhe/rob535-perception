```python
import numpy as np
import pandas as pd
from glob import glob
```

```python
train_labels = pd.read_csv('train_labels.csv',usecols=[0,3,4,5,6,7])
test_labels = pd.read_csv('test_labels.csv',usecols=[0,3,4,5,6,7])
```

```python
train_labels.head()
```

```python
def get_centroid_ground_truth(filename):
    bbox = np.fromfile(filename.replace('_image.jpg', '_bbox.bin'), dtype=np.float32)
    bbox = bbox.reshape([-1, 11])
    return bbox[0][3:6]

```

```python
# files = glob('deploy/trainval/*/*_image.jpg')
# get_centroid_ground_truth(files[0]).tolist()
train_labels['centroid_truth'] = train_labels['filename'].apply(lambda x : get_centroid_ground_truth('deploy/trainval/'+x[:36]+'/'+x[37:]))
test_labels['centroid_truth'] = test_labels['filename'].apply(lambda x : get_centroid_ground_truth('deploy/trainval/'+x[:36]+'/'+x[37:]))
```

```python
self_train_test = train_labels.append(test_labels, ignore_index=True)
self_train_test
```

```python
def find_similar_box(xmin,ymin,xmax,ymax):
    return self_train_test.apply(lambda x : (x['xmin']-xmin)**2+(x['ymin']-ymin)**2+(x['xmax']-xmax)**2+(x['ymax']-ymax)**2, axis=1).idxmin()
```

```python
from ast import literal_eval
kaggle_test = pd.read_csv('sub5result.tsv',sep='\t')
width = 1914
height = 1052
kaggle_test['ymin'] = kaggle_test['boxes'].apply(lambda x: height*literal_eval(x)[0][0])
kaggle_test['xmin'] = kaggle_test['boxes'].apply(lambda x: width*literal_eval(x)[0][1])
kaggle_test['ymax'] = kaggle_test['boxes'].apply(lambda x: height*literal_eval(x)[0][2])
kaggle_test['xmax'] = kaggle_test['boxes'].apply(lambda x: width*literal_eval(x)[0][3])
```

```python
kaggle_test.shape
```

```python
kaggle_test['centroid_guess_idx'] = kaggle_test.apply(lambda x: find_similar_box(x['xmin'],x['ymin'],x['xmax'],x['ymax']),axis=1)
```

```python
kaggle_test['centroid_guess'] = kaggle_test['centroid_guess_idx'].apply(lambda x: self_train_test['centroid_truth'][x])
```

```python
sub = pd.DataFrame()
for i in range(len(kaggle_test)):
    sub = sub.append({'guid/image/axis': kaggle_test['guid/image'][i]+'/x', 'value': kaggle_test['centroid_guess'][i][0]}, ignore_index=True)
    sub = sub.append({'guid/image/axis': kaggle_test['guid/image'][i]+'/y', 'value': kaggle_test['centroid_guess'][i][1]}, ignore_index=True)
    sub = sub.append({'guid/image/axis': kaggle_test['guid/image'][i]+'/z', 'value': kaggle_test['centroid_guess'][i][2]}, ignore_index=True)
sub
```
